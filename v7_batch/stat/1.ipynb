```python
def make_plot(file_name):
  import pandas as pd
  import matplotlib.pyplot as plt

  # Read the data from the CSV file
  df = pd.read_csv(file_name)

  # Assume the columns in your csv are named 'column1' and 'column2'
  plt.plot(df['turn'], df['gold'], marker='o')
  plt.xlabel('Column 1')
  plt.ylabel('Column 2')
  plt.title('Line Chart')
  plt.grid(True)
  plt.show()
```

```python
make_plot('1684243202.3617826.txt')
```
